BOW- most popular methods for extracting features from raw texts. The output of this model for a set of text documents is a matrix. Each column of the matrix represents a word from the vocabulary and each row corresponds to one of these text documents. Here, "vocabulary" refers to a unique set of words present in a document

BOW.jpg

Document 1: I like detective Byomkesh Bakshi.

Document 2: Byomkesh Bakshi is not a detective, he is a truth seeker.

BOW1.jpg

```
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

Document = [

'I like detective Byomkesh Bakshi.',

'Byomkesh Bakshi is not a detective, he is a truth seeker',

]

Document
```

```
# get all unique words in the corpus

vectorizer=CountVectorizer()
X = vectorizer.fit_transform(Document)
print(vectorizer.get_feature_names())
print(X.toarray())
```

The Below code is just to represent the BOW in Tabular- dataframe- It is NOT a mandatory step

```
bag_of_words_model = CountVectorizer()

print(bag_of_words_model.fit_transform(Document).todense())

bag_of_word_df = pd.DataFrame(bag_of_words_model.fit_transform(Document).todense())

bag_of_word_df.columns = sorted(bag_of_words_model.vocabulary_)

bag_of_word_df.head()
```

```
vectorizer2 = CountVectorizer(analyzer='word', ngram_range=(2, 2))
X2 = vectorizer2.fit_transform(Document)
print(vectorizer2.get_feature_names())
```

```
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

corpus = [

'Data Science is an overlap between Arts and Science',

'Generally, Arts graduates are right-brained and Science graduates are left-brained',

'Excelling in both Arts and Science at a time becomes difficult',

'Natural Language Processing is a part of Data Science'

]

corpus
```

BoW model for the 10 most frequent terms

```
bag_of_words_model = CountVectorizer(max_features=10)

print(bag_of_words_model.fit_transform(corpus).todense())

bag_of_word_df = pd.DataFrame(bag_of_words_model.fit_transform(corpus).todense())

bag_of_word_df.columns = sorted(bag_of_words_model.vocabulary_)

bag_of_word_df.head()
```
